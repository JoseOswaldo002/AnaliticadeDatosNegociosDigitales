````python
bignum = -999999999

#ingresa el primer valor
number = int(input('Introduce un numero o escribe -1 para terminar'))

while number!= -1:
    # Es el numero mas grande que el valor numero_grande?
    if number > bignum:
        bignum = number
    number = int(input('Introduce un numero o escribe -1 para terminar'))

print(f'el numero mas grande es {bignum}')
````

````python
numero = int(input('Introdice un numero'))

i = 1

while i <= numero:
    print(f'Soy el numero {i}')
    i = i + 1
print('Fin del programa')
````

````python
for i in range(10):
    print(f'El valor de i es {i}')
````

### Las sentencias `Brake` y `continue`

`break` - Sale del bucle inmediatamente, e incondicionalmente termina la operacin del bucle, el programa empieza a ejecutar la instruccin mas cercana despus del cuerpo del bucle

`continue` - Se comporta como si el programa hubiera llegado repentinamente al final del cuerpo; el siguiente turno se inicia y la expresin de condicin se prueba de inmediato

````python
# breake - ejemplo

print('La instruccion break')
for i in range(1,6):
    if i == 3:
        break
    print('Dentro del bucle')
print('Fuera del bucle')
````

````python
# continue

print ('\nLa instruccion continue: ')
for i in range(1,6):
    if i == 3:
        continue
    print(f"Dentro del bucle {i}")
print('Fuera del bucle')
````

## listas en Accin

Imagina que necesitas reorganizar los elementos de una lista,es decir, revertir el orden de los elementos: el primero y el quinto, asi como el
segundo y el cuarto elemento sern intercambiados. El tercero permanecera intacto.

Pregunta:  Como se pueden intercambiar los valores de dos variables?

````python
variable_1 = 1
variable_2 = 2

variable_1, variable_2 = variable_2, variable_1

print(f'El valor de la variable_1 es: {variable_1}\nEl valor de la variable_2 es {variable_2} ')
````

Teniendo una lista llamada mi_lista con valores, 10,1,8,3,5, intercambiar el 10 por el 5 v viceversa

````python
mi_lista = [10,1,8,3,5]

mi_lista[0],mi_lista[4] = mi_lista[4], mi_lista[0]
mi_lista[1],mi_lista[3] = mi_lista[3], mi_lista[1]

print(mi_lista)
````
